# Binomial Distribution

Binomial distribution is a discrete probability distribution that gives two results in an experiment, recorded as a success (1) or failure (0). The binomial distribution helps us understand the probabilities associated with a series of independent trials, each with the same probability of success.

In Binomial distribution there are two key parameters; n and p.
n = the number of times an experiment runs
p = the probability of any one outcome.

# Getting our hands slightly dirty

```python
# Importing necessary modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 12 # number of trials
p = 0.167 #probability of getting a 4

x_values = np.arange(0, n+1)

probabilities = [np.math.comb(n, k) * (p**k) * ((1-p)**(n-k)) for k in x_values]

# Plot the probability distribution graph
plt.bar(x_values, probabilities, align='center', alpha=0.5)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution: n={}, p={}'.format(n, p))
plt.xticks(x_values)
plt.grid(True)
plt.show()
```
